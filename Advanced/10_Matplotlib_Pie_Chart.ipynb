```python
#Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datasets import load_dataset
import ast

# Loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data cleanup
df.job_posted_date = pd.to_datetime(df.job_posted_date)
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df.job_work_from_home.value_counts().plot(kind='pie', startangle =90, autopct='%1.1f%%')
plt.title('Ratio of remote jobs')
plt.ylabel('')
plt.show()
```

## Using subplots to plot multiple pie chart

```python
df[['job_work_from_home','job_no_degree_mention','job_health_insurance']]
```

```python
dict_column = {'job_work_from_home': 'Work from home', 'job_no_degree_mention': 'Degree request', 'job_health_insurance': 'Health insurance covered'}
```

```python
fig, ax = plt.subplots(1,3)

for i, (column, title) in enumerate(dict_column.items()): # .items is used to get the key-value pair from the dictionary
    ax[i].pie(df[column].value_counts(), startangle=90, autopct='%1.1f%%', labels =['False', 'True'])
    ax[i].set_title(title)


fig.tight_layout()
```

```python
fig, ax = plt.subplots(1,3)

for i, (column, title) in enumerate(dict_column.items()): # .items is used to get the key-value pair from the dictionary
    df[column].value_counts().plot(kind='pie', ax=ax[i], startangle =90, autopct='%1.1f%%')
    ax[i].set_title(title, fontweight = 'bold')  # set the title for individual plot and make the title bold
    ax[i].set_ylabel('')

fig.tight_layout()
```
